```julia
using CSV
using DataFrames
using Plots
using StatsPlots
```

## Wage data

```julia
wageData = CSV.File("dataset/Wage.csv") |> DataFrame;
```

```julia
wage = wageData.wage
age = wageData.age
year = wageData.year
edu = wageData.education

p1 = scatter(age, wage, xlabel="Age", ylabel="Wage")
p2 = scatter(year, wage, xlabel="Year")
p3 = boxplot(edu, wage, xlabel="Education Level", ylabel="Wage")
plot(p1, p2, p3, layout = 3, legend = false)
```
